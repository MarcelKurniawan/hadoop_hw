```python
import os
import requests
import csv

# Define the base URL for the API endpoint
BASE_URL = "https://pokeapi.co/api/v2/ability/"

# Function to fetch ability data from the API
def fetch_ability_data(ability_id):
    response = requests.get(BASE_URL + str(ability_id))
    if response.status_code == 200:
        data = response.json()
        effects = data['effect_entries']
        result = []
        for effect in effects:
            result.append({
                "id": data['id'],
                "pokemon_ability_id": ability_id,
                "effect": effect['effect'],
                "language": effect['language']['name'],
                "short_effect": effect['short_effect']
            })
        print(f"Data pulled for ability ID: {ability_id}")
        return result
    else:
        return []

# Function to save data to a CSV file
def save_to_csv(start_id, end_id, data):
    directory = "csv"
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    filename = os.path.join(directory, f"result_{start_id}_{end_id}.csv")
    with open(filename, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=["id", "pokemon_ability_id", "effect", "language", "short_effect"])
        writer.writeheader()
        for row in data:
            writer.writerow(row)
    
    print(f"CSV file created: {filename}")

# Loop through ability IDs and save data in batches of 100
all_data = []
for ability_id in range(1, 1000):
    all_data.extend(fetch_ability_data(ability_id))
    if ability_id % 100 == 0:
        start_id = ability_id - 99
        end_id = ability_id
        save_to_csv(start_id, end_id, all_data)
        all_data = []

# Save any remaining data
if all_data:
    start_id = (ability_id // 100) * 100 + 1
    end_id = ability_id
    save_to_csv(start_id, end_id, all_data)
```
